```python
import pandas as pd 
import os
import pickle as pkl
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,roc_auc_score,confusion_matrix
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
path="C:\\Users\\andyr\\Documents\\file\\side\\projects\\ufc"
x_train=pd.read_csv(os.path.join(path,"x_train.csv"))
x_test=pd.read_csv(os.path.join(path,"x_test.csv"))
y_train=pd.read_csv(os.path.join(path,"y_train.csv"))
y_test=pd.read_csv(os.path.join(path,"y_test.csv"))
```

```python
#tried decision tree 
model1=DecisionTreeClassifier(max_depth=3,class_weight="balanced")

model1.fit(x_train,y_train)

preds=model1.predict(x_test)


print(classification_report(y_test,preds))

preds_train=model1.predict(x_train)

print(classification_report(y_train,preds_train))

print(confusion_matrix(y_train,preds_train))

print(confusion_matrix(y_test,preds))

print(roc_auc_score(y_test,preds))
```

```python
#tried logistic regression
model2=LogisticRegression(class_weight="balanced",max_iter=10000,solver="lbfgs")

model2.fit(x_train,y_train)

preds=model2.predict(x_test)

print(classification_report(y_test,preds))

preds_train=model2.predict(x_train)

print(classification_report(y_train,preds_train))
print(roc_auc_score(y_train,preds_train))

print(confusion_matrix(y_train,preds_train))

print(confusion_matrix(y_test,preds))
print(roc_auc_score(y_test,preds))
```

```python
#tried random forest again with different parameters
model3=RandomForestClassifier(n_estimators=1000,max_depth=4)

model3.fit(x_train,y_train)

preds=model3.predict(x_test)

print(classification_report(y_test,preds))

preds_train=model3.predict(x_train)
print(roc_auc_score(y_train,preds_train))

print(classification_report(y_train,preds_train))
print(confusion_matrix(y_train,preds_train))

print(confusion_matrix(y_test,preds))
print(roc_auc_score(y_test,preds))
```

```python
with open("model1.pkl","wb") as f:
    pkl.dump(model1,f)

with open("model2.pkl","wb") as f:
    pkl.dump(model2,f)
    
with open("model3.pkl","wb") as f:
    pkl.dump(model3,f)
    
```
