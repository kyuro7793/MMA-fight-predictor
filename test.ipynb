```python
import numpy as np
import pandas as pd
import os
import pickle as pkl
```

```python
path="C:\\Users\\andyr\\Documents\\file\\side\\projects\\ufc\\datasets"
testdata=pd.read_csv(os.path.join(path,"upcoming-event.csv"))
```

```python
with open("model1.pkl","rb") as f:
    model1 = pkl.load(f)

with open("model2.pkl","rb") as f:
    model2 = pkl.load(f)
    
with open("model3.pkl","rb") as f:
    model3 = pkl.load(f)
```

```python
with open("selected_features.pkl","rb") as f:
    select_feats=pkl.load(f)
```

```python
with open("encoding.pkl","rb") as f:
    encoding=pkl.load(f)
```

```python
os.listdir()
```

```python
with open("median.pkl","rb") as f:
    median = pkl.load(f)

with open("mode.pkl","rb") as f:
    mode = pkl.load(f)
```

```python
test=testdata[select_feats]
```

```python
test.finish=test.finish.astype("object")
```

```python
for i in test.columns:
    if test[i].isnull().sum()>0:
        if test[i].dtype=="object":
            test[i]=test[i].fillna(mode[i])
        else:
            test[i]=test[i].fillna(median[i])
        
```

```python
for i in test.columns:
    if test[i].dtype=="object":
        test[i]=test[i].apply(lambda x: encoding[i][x])
```
