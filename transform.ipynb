# Feature encoding

```python
encoding={}
for i in data.columns:
    
   
    if i=="Winner":
        continue
    if data[i].dtype=="object":
        enc=data.groupby(i).size()/data.shape[0]
        encoding[i]=enc

        data[i]=data[i].apply(lambda x: enc[x])
```

```python
data.Winner=data.Winner.apply(lambda x: 0 if x=="Red" else 1)
```
