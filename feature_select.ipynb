```python
import pandas as pd
import os
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
import pickle as pkl
```

```python
path="C:\\Users\\andyr\\Documents\\file\\side\\projects\\ufc"
data=pd.read_csv(os.path.join(path,"transformed_data.csv"))
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,roc_auc_score,confusion_matrix
```

```python
X=data.drop("Winner",axis=1)
```

```python
y=data["Winner"]
```

```python
y.value_counts()
```

```python
rfe=RFE(RandomForestClassifier())
```

```python
rfe.fit(X,y)
```

```python
select_feats=[]
for name,selected in zip(X.columns,rfe.support_):
    if selected==True:
        select_feats.append(name)
```

```python
X=X[select_feats]
```

```python
with open("selected_features.pkl","wb") as f:
    pkl.dump(select_feats,f)
```

```python
X.shape
```

```python
X.to_csv("X_feature_selected.csv",index=False)
```

```python
y.to_csv("y.csv",index=False)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=1)
```

```python
x_train.to_csv("x_train.csv",index=False)
x_test.to_csv("x_test.csv",index=False)
y_train.to_csv("y_train.csv",index=False)
y_test.to_csv("y_test.csv",index=False)
```
